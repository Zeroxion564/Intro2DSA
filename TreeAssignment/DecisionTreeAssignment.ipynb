```python
import pandas as pan
import numpy as np

data = np.array(pan.read_csv("Tree-classification.csv"))

def pricePrediction(house):
    if house[1]>160:
        if house[3]>3:
            if house[2]<10: return 510
            else: return 382
        else:
            if house[2]<40: return 335
            else: return 312
    else:
        if house[2]<25:
            if house[3]>2: return 341
            else: return 258
        else:
            if house[1]>100: return 285
            else: return 217
            
for i in data:
    print(pricePrediction(i))
```
