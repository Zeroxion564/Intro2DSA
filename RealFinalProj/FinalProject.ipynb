```python
%matplotlib inline
import numpy as np
import pandas as pnd
import matplotlib as plt
import seaborn as sns; sns.set()

sns.set(style="darkgrid")
```

The first Database is the New York income tax data by tax brackets.

```python
#Load Data
df = pnd.read_excel("State2015.xlsx")

#Clean Data
#Need to isolate the "All Places" and "All Filer" rows in their own summary.
#The summary of all of New York is what is needed most.

#First, isolate the important summary rows.
#These rows can be identified by the entries by the "NYS Residency Status", "Place of Residence", "County", and "State" rows.
#The rows summerizing everything will have the values "All Filers", "All Places", "All Places", and "All" res[ectively
#By creating a new Dataframe that shows where they overlap we can pull the row indexes out and create a smaller summary table

#Creating the temporary Dataframe
tempDF = pnd.DataFrame(data=(df["NYS Residency Status"]=="All Filers") &
                            (df["Place of Residence"]=="All Places") & 
                            (df["County"]=="All Places") & 
                            (df["State"]=="All"), 
                       columns=["Bool"])
#Isolating the rows we want
rows = tempDF[tempDF["Bool"]==True].index
#Creating the uncleaned TaxSummary table
tempDF = df.iloc[rows]

#Cleaning the TaxSummary means selecting the columns that are needed most.
#FOr now, that is the "NY Adjusted Gross Income Range (Fed.Col)", "Number of Returns",
#"Federal Adjusted Gross Income", "Taxable Income"
TaxSummary = tempDF[["NY Adjusted Gross Income Range (Fed.Col)", "Number of Returns", "Federal Adjusted Gross Income", "Taxable Income"]]

#For a point of reference, there should also be a number representing the percentage of returns in each tax bracket.
TaxSummary = TaxSummary.assign(Pop_Percent=pnd.Series((100*TaxSummary["Number of Returns"]/TaxSummary["Number of Returns"].max()), index=TaxSummary.index))

#Restructuring and renaming.
#For ease of reading, it would be best to rename several columns at this point.
#In addition, reordering the columns and resseting the index would be helpful.

#Remove the underscore in "Pop_Percent" and remove some excess words from
#"NY Adjusted Gross Income Range (Fed.Col)" and "Federal Adjusted Gross Income"
TaxSummary = TaxSummary.rename(index=str, columns={"NY Adjusted Gross Income Range (Fed.Col)": "Adjusted Gross Income Range",
                                      "Federal Adjusted Gross Income": "Adjusted Gross Income",
                                      "Pop_Percent": "Pop Percent"})

#reset inedex
TaxSummary = TaxSummary.reset_index(drop=True)

#reorder columns
TaxSummary = TaxSummary[["Adjusted Gross Income Range", "Number of Returns", "Pop Percent", "Adjusted Gross Income", "Taxable Income"]]
```
